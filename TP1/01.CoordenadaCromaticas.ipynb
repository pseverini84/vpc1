```python
#Si queremos que las imágenes sean mostradas en una ventana emergente quitar el inline
%matplotlib inline

# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt
```

```python
img = cv.imread('CoordCrom_1.png')
imgRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)

R,G,B=cv.split(imgRGB)

R=R/(np.sum(imgRGB, axis=2)+0.001)
G=G/(np.sum(imgRGB, axis=2)+0.001)
B=B/(np.sum(imgRGB, axis=2)+0.001)

img_crom=cv.merge((R,G,B))

img.astype('uint8')

plt.imshow(img_crom)
plt.show()

img = cv.imread('CoordCrom_2.png')
imgRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)

R,G,B=cv.split(imgRGB)

R=R/(np.sum(imgRGB, axis=2)+0.001)
G=G/(np.sum(imgRGB, axis=2)+0.001)
B=B/(np.sum(imgRGB, axis=2)+0.001)

img_crom=cv.merge((R,G,B))

img.astype('uint8')

plt.imshow(img_crom)
plt.show()

img = cv.imread('CoordCrom_3.png')
imgRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)

R,G,B=cv.split(imgRGB)

R=R/(np.sum(imgRGB, axis=2)+0.001)
G=G/(np.sum(imgRGB, axis=2)+0.001)
B=B/(np.sum(imgRGB, axis=2)+0.001)

img_crom=cv.merge((R,G,B))

img.astype('uint8')

plt.imshow(img_crom)
plt.show()
```
